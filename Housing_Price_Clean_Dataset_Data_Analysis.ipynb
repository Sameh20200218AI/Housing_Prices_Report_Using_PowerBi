### Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Read Dataset

```python
data = pd.read_csv('Housing.csv')
data
```

```python
print(data.describe())
```

```python
# Checking Null values
data.isnull().sum()*100/data.shape[0]
```

```python
data.shape
```

```python
data
```

```python
# Renaming columns 
data = data.rename(columns={'price': 'Price',
                             'area': 'Area',
                             'bedrooms':'Bedrooms',
                             'bathrooms':'Bathrooms',
                             'stories':'Stories',
                             'mainroad' :'Main Road',
                             'guestroom' : 'Guest Room',
                             'basement':'Basement',
                             'hotwaterheating': 'Hot Water Heating',
                             'airconditioning':'Airconditioning',
                             'parking':'Parking',
                             'prefarea':'Prefare Area',
                             'furnishingstatus':'Furnishing Status' })
data
```

```python
data['Main Road']=data['Main Road'].replace({'yes':'Yes','no':'No'})
data['Guest Room']=data['Guest Room'].replace({'yes':'Yes','no':'No'})
data['Basement']=data['Basement'].replace({'yes':'Yes','no':'No'})
data['Hot Water Heating']=data['Hot Water Heating'].replace({'yes':'Yes','no':'No'})
data['Airconditioning']=data['Airconditioning'].replace({'yes':'Yes','no':'No'})
data['Prefare Area']=data['Prefare Area'].replace({'yes':'Yes','no':'No'})
data['Furnishing Status']=data['Furnishing Status'].replace({'furnished':'Furnished','semi-furnished':'Semi-Furnished','unfurnished':'Unfurnished'})
data
```

```python
fig = plt.figure(figsize=(6,8))
ax1 = fig.add_subplot(1,2,1)
ax1.boxplot(data['Area'])
ax1.set_title('Box Plot for Area ')

ax2 = fig.add_subplot(2,2,2)
ax2.boxplot(data['Price'])
ax2.set_title('Box Plot Price')

plt.tight_layout()
plt.show()
```

### Remove Outliers

```python
#remove outliers from Area and Price Features

#Q1 is the 25% and Q3 is 75% of data, Q2 is 50% of data

Q1 = data.Area.quantile(0.25)
Q3 = data.Area.quantile(0.75)
IQR = Q3 - Q1
data = data[(data.Area >= Q1 - 1.5*IQR) & (data.Area <= Q3 + 1.5*IQR)] # drop rows that include outliers


Q1 = data.Price.quantile(0.25)
Q3 = data.Price.quantile(0.75)
IQR = Q3 - Q1
data = data[(data.Price >= Q1 - 1.5*IQR) & (data.Price <= Q3 + 1.5*IQR)] # drop rows that include outliers

data.shape
```

### Visualization

```python
fig = plt.figure(figsize=(9,11))
ax1 = fig.add_subplot(5,2,1)
ax1.scatter(data['Area'],data['Price'],marker='*' ,color='green')
ax1.set_title('Area VS Price ')

ax2 = fig.add_subplot(5,2,2)
ax2.scatter(data['Bedrooms'],data['Price'],marker='*' ,color='red')
ax2.set_title('Bedrooms VS Price ')

ax3 = fig.add_subplot(5,2,3)
ax3.scatter(data['Bathrooms'],data['Price'],marker='*' ,color='cyan')
ax3.set_title('Bathrooms VS Price ')

ax4 = fig.add_subplot(5,2,4)
ax4.scatter(data['Stories'],data['Price'],marker='*' ,color='brown')
ax4.set_title('Stories VS Price')


plt.tight_layout()
plt.show()
```

```python
# Define the number of bins for 'Area Quality' and 'Price Quality' features

n_bins = 4  # Number of categories for area quality
bins_area = np.linspace(data['Area'].min(), data['Area'].max(), n_bins + 1)

# Define labels for the bins
labels_area = ['Very Low Area', 'Low Area', 'High Area', 'Very High Area']

# Create the 'Area Quality' column
data['Area Quality'] = pd.cut(data['Area'], bins=bins_area, labels=labels_area, include_lowest=True)


bins_price = np.linspace(data['Price'].min(), data['Price'].max(), n_bins + 1)

labels_price = ['Very Low Price', 'Low Price', 'High Price', 'Very High Price']

# Create the 'Price Quality' column
data['Price Quality'] = pd.cut(data['Price'], bins=bins_price, labels=labels_price, include_lowest=True)
data

```

```python
# Save the Dataframe to a CSV file
data.to_csv('Housing_Price_Cleaned.csv', index=False)
```
